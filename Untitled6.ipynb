```python
import matplotlib.pyplot as plt

# Define the points
x_coordinates = [1, 2, 3, 4]
y_coordinates = [2, 4, 6, 8]

# Create the plot
plt.plot(x_coordinates, y_coordinates, color='red', linestyle='--', marker='o', label='Line Graph')

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Graph of Points')

# Add a legend
plt.legend()

# Add grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the points
x_coordinates = [1, 2, 3, 4]
y_coordinates = [2, 4, 6, 8]

# Create the bar graph
plt.bar(x_coordinates, y_coordinates, color='green', edgecolor='white', linestyle='-', linewidth=2, label='Bar Graph')

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Bar Graph of Points')

# Add a legend
plt.legend()

# Add grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```
