```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bokeh.plotting import figure, output_notebook, show
from bokeh.io import push_notebook

data = pd.read_csv("/content/environmental_data.csv")

plt.figure(figsize=(10, 6))
sns.lineplot(data=data)
plt.title('Line Graph of Temperature, Humidity, and Pressure')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(data.columns)
plt.show()

plt.figure(figsize=(8, 8))
pressure_data = data['Pressure']
labels = data.index
plt.pie(pressure_data, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Pressure')
plt.axis('equal')
plt.show()

output_notebook()

p = figure(x_range=data.columns.tolist(), height=250, title="Bar Graph of Average Values")
p.vbar(x=data.columns.tolist(), top=data.mean(), width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p, notebook_handle=True)

print("\nAnalysis Summary:")
print("\nLine Graph:")
print("This line graph shows the trends of Temperature, Humidity, and Pressure over the dataset indices. The fluctuations in each variable can be observed clearly.")

print("\nPie Chart:")
print("The pie chart represents the distribution of Pressure values across different indices. Each segment shows the proportion of the Pressure at each data point.")

print("\nBar Graph:")
print("The bar graph displays the average values of Temperature, Humidity, and Pressure. This provides a quick comparison of the central tendency for each variable in the dataset.")
```
